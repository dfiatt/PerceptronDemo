```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
from IPython.display import display, Markdown

# Definición exacta de la clase Perceptron que usaste para entrenar
class Perceptron:
    def __init__(self, input_size, learning_rate=0.01, epochs=10):
        self.w = np.zeros(input_size)
        self.b = 0
        self.lr = learning_rate
        self.epochs = epochs

    def activation(self, z):
        return 1 if z > 0 else 0

    def predict(self, x):
        z = np.dot(self.w, x) + self.b
        return self.activation(z)

# Cargar datos desde CSV para visualización
clientes = []
with open("data.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        clientes.append({
            "age": float(row['age']),
            "freq": float(row['freq']),
            "spent": float(row['spent']),
            "label": int(row['label'])
        })

# Convertir a numpy array para predicciones
X = np.array([[c["age"], c["freq"], c["spent"]] for c in clientes])
y = np.array([c["label"] for c in clientes])

# Cargar modelo entrenado
with open("perceptron_model.pkl", "rb") as f:
    perceptron = pickle.load(f)

# Mostrar explicación del gráfico
display(Markdown("### Interpretación del gráfico"))
display(Markdown(
"""
- Cada punto representa un cliente.
- El eje X muestra la edad.
- El eje Y muestra el monto promedio gastado.
- Puntos **verdes**: clasificación correcta por el perceptrón.
- Puntos **rojos**: clasificación incorrecta.
"""
))

# Graficar Edad vs Monto Promedio con colores según aciertos del perceptrón
fig, ax = plt.subplots(figsize=(8,6))
for i, point in enumerate(X):
    color = "green" if perceptron.predict(point) == y[i] else "red"
    ax.scatter(point[0], point[2], color=color, s=100)
ax.set_xlabel("Edad")
ax.set_ylabel("Monto Promedio")
ax.set_title("Perceptrón - Clasificación de Clientes (verde: correcto, rojo: error)")
ax.grid(True)
plt.show()

# Pedir al usuario un precio específico para filtrar clientes receptivos
precio_producto = float(input("Ingresa el precio del producto para filtrar clientes receptivos: "))

# Filtrar clientes receptivos para ese precio (predicción = 1 y spent >= precio_producto)
clientes_receptivos = []
for c in clientes:
    x = np.array([c["age"], c["freq"], c["spent"]])
    if perceptron.predict(x) == 1 and c["spent"] >= precio_producto:
        clientes_receptivos.append(c)

# Mostrar la lista en tabla ordenada por monto gastado descendente
if clientes_receptivos:
    df_receptivos = pd.DataFrame(clientes_receptivos)
    df_receptivos = df_receptivos.sort_values(by="spent", ascending=False)
    display(Markdown(f"### Clientes receptivos a la promoción para productos de precio >= {precio_producto}"))
    display(df_receptivos)
else:
    display(Markdown(f"### No se encontraron clientes receptivos para productos de precio >= {precio_producto}"))
```
